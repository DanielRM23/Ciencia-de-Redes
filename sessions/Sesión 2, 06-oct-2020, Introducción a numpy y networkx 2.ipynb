# Funciones en python.

#### Se define un a función para poder hacer un anillo hasta tres primeros vecinos.
#### El argumento de la función es el número de nodos.

```python
import networkx as nx
import numpy as np


def Anillo(N): 
    
    Anillo = nx.Graph()

    #Numero de nodos
    #N = 15

    for i in range(N):
        Anillo.add_node(i)

    lista_nodos = list(Anillo.nodes)

    for i in range(N):
        #Enlace a primeros vecinos.
        Anillo.add_edge(lista_nodos[i-1] , lista_nodos[i])
        #Enlace a segundos vecinos.
        Anillo.add_edge(lista_nodos[i-2] , lista_nodos[i])
        #Enlace a terceros vecinos.
        Anillo.add_edge(lista_nodos[i-3] , lista_nodos[i])
    
    #nx.draw(Anillo)
    nx.draw_circular(Anillo, with_labels = True)
    
    return Anillo
```

#### Con la función se guarda la gráfica, de esta manera podemos nombrarla como una variable para poder trabajar con ésta.

```python
anillo_20 = Anillo(20)
```

# G.add_nodes_from( )
# G.add_edges_from( )


#### Sirve para agregar nodos y vértices desde una forma más rápida y concisa.

#### Agregamos nodos desde un arreglo (array) hecho por numpy.

```python
import networkx as nx
import numpy as np

#Se define una función que regresa una red lineal.
def Red_lineal(N):
    
    #Se define el objeto graph
    red_lineal = nx.Graph()
    
    #Se definen los nodos con el comando arange.
    nodos = np.arange(N)
    
    #En esta línea se agregan los nodos con la función descrita anteriormente.
    red_lineal.add_nodes_from(nodos)
    #De forma análoga se agregan las aristas, pero se utiliza la función zip.
    red_lineal.add_edges_from(zip(nodos[:-1], nodos[1:]))
    
    #Con esta linea se dibuja la red.
    nx.draw(red_lineal, with_labels = True)
    
    return red_lineal
```

```python
Red_lineal(20)
```

#### Esta función sirve para recorrer en un número $n$ el array.

#### En el siguiente ejemplo se están recorriendo los elementos un lugar, de esta forma el último elemento del array pasa a ser el primer elemento.

```python
import numpy as np

N = 20
nodos = np.arange(N)
np.roll(nodos, 1)

list(zip(nodos, np.roll(nodos, 1) ) )
```

```python
import networkx as nx
import numpy as np

def Circular(N):

    #Se define el objeto graph
    circular = nx.Graph()
    
    #Se definen los nodos.
    nodos = np.arange(N)    

    #Se agregan los nodos desde el comando realizado con numpy.
    circular.add_nodes_from(nodos)
    #Se agregan las aristas utilizando el comando para llamarlos desde numpy.
    #En este caso s utiliza, en conjunto la funcion zip y roll, lo anterior se muestra en la celda de arriba.
    circular.add_edges_from( zip(nodos, np.roll(nodos,1)) )
    
    #Se dibuja la función.
    nx.draw(circular,with_labels=True)

    return circular
```

```python
Circular(20)
```

# Red de anillos.

```python
import networkx as nx
import numpy as np

def Red_anillo(N):

    anillo = nx.Graph()

    nodos = np.arange(N)

    anillo.add_nodes_from(nodos)

    anillo.add_edges_from( zip(nodos, np.roll(nodos,1)) )
    anillo.add_edges_from( zip(nodos, np.roll(nodos,2)) )
    anillo.add_edges_from( zip(nodos, np.roll(nodos,3)) )

    nx.draw_circular(anillo,with_labels=True)

    return anillo
```

```python
Red_anillo(10)
```

# Red Completa

```python
import numpy as np
import networkx as nx

#Número de nodos
N = 4

completa = nx.Graph()

#Aquí se fija primero este valor, después entra en el cíclo para la j.
for i in range(N):
    #Se pone hasta i para que no se repitan 2 veces los enlaces.
    for j in range(i):
        completa.add_edge( i , j)
        
nx.draw_circular(completa, with_labels=True)        
```

# nx.from_numpy_array(A)

#### Se utiliza esta función para poder hacer la representación de una red en forma gráfica.

#### En el siguiente ejemplo se define una matriz de adyacencias de alguna red, después se almacena en un array, como último paso, este último es usado como argumento en la función *nx.from_numpy_array* para así trabajar con la información que nos da la matriz de adyacencias.

```python
#Se define una lista cuyos elementos son listas.
matriz = [[0,1,1,1],
 [1,0,0,0],
 [1,0,0,1],
 [1,0,1,0]]

#Se convierte en un array la matriz para poder trabajar con ella de una mejor manera.
A = np.array(matriz)
#Aquí se está extrayendo la información de la matriz de adyacencias.
G = nx.from_numpy_array(A)
#Se dibuja la red.
nx.draw_circular(G, with_labels=True)
```

```python
import networkx as nx
import numpy as np

def Circular(N):

    #Se define el objeto graph
    circular = nx.Graph()
    
    #Se definen los nodos.
    nodos = np.arange(N)    

    #Se agregan los nodos desde el comando realizado con numpy.
    circular.add_nodes_from(nodos)
    #Se agregan las aristas utilizando el comando para llamarlos desde numpy.
    #En este caso s utiliza, en conjunto la funcion zip y roll, lo anterior se muestra en la celda de arriba.
    circular.add_edges_from( zip(nodos, np.roll(nodos,1)) )
    
    #Se dibuja la función.
    #nx.draw(circular,with_labels=True)

    return circular
```

#### Esta es la visualización de la red.

```python
nx.draw(Circular(10), with_labels= True)
```
