```python
import networkx as nx
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
#Con esta se puede ver la modularidad de las redes.
from networkx.algorithms.community import greedy_modularity_communities


#Este es el nombre del archivo.
archivo = "bn-mouse_visual-cortex_2.edges"

#Importamos el archivo usando pandas.
datos = pd.read_csv(archivo, 
                    #rows = 10, 
                    header = None, 
                    sep = " ")

#En esta lnea se guardan los datos de pandas en una variable del tipo networkx.
G = nx.from_pandas_edgelist(datos, source = 0, target = 1 )

comunidades = list(greedy_modularity_communities(G))

diccionario = {}

for i, comunidad in enumerate(comunidades): #Visita las comunidades
    
    for nodo in comunidad: # Visita los nodos
        #print(nodo,i)
        diccionario[nodo] = i
        
colors = np.array([diccionario[i] for i in G])

plt.figure(figsize=[8,8])
nx.draw_kamada_kawai(G, #pos = position, 
                    node_color = colors,
                    cmap = "nipy_spectral"
                    )
```

```python
#list(nx.connected_components(G))
```

```python
max_com= max([c for c in nx.connected_components(G)], key=len)
Gm = G.subgraph(max_com)
plt.figure(figsize=[8,8])
nx.draw_kamada_kawai(Gm, #pos = position, 
                    node_color = colors,
                    cmap = "nipy_spectral"
                    )
```

```python
componentes = list(nx.connected_components(G))
for i,componente in enumerate(componentes): 
    Gm = G.subgraph(componente)
    print("Componente", i)
    nx.draw(Gm)
    plt.show()
```
