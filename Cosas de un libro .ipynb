```python
import networkx as nx 
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
G = nx.Graph()

for i in range(1,10):
    G.add_node(i)
    G.add_edge(i,i+1)
    
nx.draw(G, with_labels = True)

for n in G.nodes:
    print(n, G.neighbors(n))
```

```python
B = nx.complete_bipartite_graph(4,4)
nx.draw_shell(B, with_labels=True)

N = B.number_of_nodes()
L = B.number_of_edges()
L_max = 0.5*(N)*(N-1)
density = L/L_max
#density

nx.density(B)
```

```python
C = nx.cycle_graph(4)
nx.draw_shell(C)
```

```python
S = nx.star_graph(6)
nx.draw(S, with_labels=True)
```

```python
sub = nx.subgraph(S,(0,1,2))
nx.draw(sub, with_labels = True)
```
