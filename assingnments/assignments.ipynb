```python
#Estas son todas las libreras que se van a utilizar.
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd

#Este es el nombre del archivo.
archivo = "bn-mouse_visual-cortex_2.edges"

#Importamos el archivo usando pandas.
datos = pd.read_csv(archivo, 
                    #rows = 10, 
                    header = None, 
                    sep = " ")

#En esta lnea se guardan los datos de pandas en una variable del tipo networkx.
G = nx.from_pandas_edgelist(datos, source = 0, target = 1 )

#Definimos la centralidad de nodo.
degree = nx.degree_centrality(G)

#El valor de las centralidades lo anexamos a una variable para poder usarla despus.
#Se utiliza una lista de compresn. Se toman los elementos que pertenecen a G, si no se hace esto
#se pude estar dando otra interpretacin a los valores.
Degree_centrality_array = np.array( [degree[i] for i in G] )


#Definimos la centralidad de Katz y a esta le asignamos la variable para los colores.
Katz = nx.katz_centrality(G)
Katz_centrality_array= np.array( [Katz[i] for i in G] ) 


#Hacemos la imagen ms grande.
plt.figure(figsize = [8,8])
#Dibujamos la red.
nx.draw_kamada_kawai(G,
       node_size = 10000*Degree_centrality_array,
       node_color = Katz_centrality_array,
       #cmap = "spring",
       )

#Para guardar la figura.
#plt.savefig('Mouse_visual_cortex', dpi = 200)
 
#Definimos la centralidad de cercana.
closeness = nx.closeness_centrality(G)
#Asignamos esta variable para guardar los datos anteriores.
Closeness_centrality_array = np.array([closeness[i] for i in G])

between = nx.betweenness_centrality(G)
Between_centrality_array= np.array( [between[i] for i in G] ) 

#Se grafican los resultados.
plt.figure(figsize=[15,5])
plt.grid()
plt.title("Medidas de Centralidad.")
plt.plot(Degree_centrality_array, label= "Centralidad de nodo.")
plt.plot(Closeness_centrality_array, label= "Centralidad de cercana.")
plt.plot(Between_centrality_array, label= "Centralidad de intermediacin.")
plt.plot(Katz_centrality_array, label= "Centralidad de Katz.")
#plt.plot(e, label= "Centralidad de eigenvector.")
plt.legend()
plt.show()
```
