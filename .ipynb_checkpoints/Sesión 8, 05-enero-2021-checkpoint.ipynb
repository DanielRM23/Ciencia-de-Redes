```python
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
import pandas as pd
```

```python
archivo = 'bn-mouse_visual-cortex_2.edges'
```

```python
ruta = pd.read_csv(archivo,
                  #nrows=10,
                  header = None,
                  sep = " ")
```

```python
import networkx as nx
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Asumiendo que 'ruta' es un DataFrame con las columnas correspondientes
G = nx.from_pandas_edgelist(ruta, source=0, target=1)

# Se almacena el grado de cada nodo en un diccionario.
diccionario = dict(G.degree)

# Se convierte en lista los grados de los nodos.
grado = [diccionario[i] for i in G]

# Se genera la probabilidad de cada nodo.
probabilidad = [grado[i]/sum(grado) for i in range(len(grado))]

# Se toma una probabilidad al azar.
proba_enlace = [np.random.choice(probabilidad) for _ in range(len(grado))]

# Se hace un diccionario con los valores de las probabilidades con sus respectivos nodos.
dictionary = {i: probabilidad[i] for i in range(len(grado))}

# Se agregan los nodos a la red.
for i in dictionary.keys():
    if proba_enlace[i] == dictionary[i]:
        G.add_edge(0, i)

# Dibujar el grafo.
plt.figure(figsize=[10, 10])
nx.draw_kamada_kawai(G, with_labels=True)
plt.show()
```

```python
def enlaces_aleatorios(N, probabilidad):
  for i in range(N):
    for j in range(i):
      if np.random.rand() < probabilidad:
        yield i,j

def red_aleatoria(N, probabilidad):
  G = nx.Graph()
  G.add_nodes_from(range(N)) # agrega todos los nodos
  
  probabilidad = []
  for i in grado:
    proba = grado[i]/sum(grado)
    probabilidad.append(proba)

  G.add_edges_from( enlaces_aleatorios(N, probabilidad) ) # agrega los enlaces obedeciendo la probabilidad p
  
  return G
```
